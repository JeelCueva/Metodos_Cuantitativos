## Sección 5.1 — La Función Exponencial
### Ejemplo 5.1 — Familia de funciones exponenciales (crecientes y decrecientes)

**Propiedades de** $f(x)=a^x$:
- Si $a>1$: crecimiento exponencial
- Si $0<a<1$: decaimiento exponencial
- El número de Euler: $e = \lim_{n\to\infty}\left(1+\frac{1}{n}\right)^n \approx 2.71828$


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 5.1 — Familia de funciones exponenciales
# ═══════════════════════════════════════════════════════════════
x = np.linspace(-3, 3, 400)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Capítulo 5 — Familia de Funciones Exponenciales', fontsize=14, fontweight='bold')

# ── Panel izquierdo: crecientes ──────────────────────────────
ax = axes[0]
bases_crec = [1.5, np.e, 2, 3, 5]
colores = [C_AZUL, C_ROJO, C_VERDE, C_NARANJA, C_MORADO]
for a, col in zip(bases_crec, colores):
    y = a**x
    label = f'e^x (e≈2.718)' if a == np.e else f'{a}^x'
    ax.plot(x, y, color=col, lw=2, label=label)

ax.set_ylim(-0.5, 10)
ax.axhline(0, color='black', lw=0.8, ls='--')
ax.axvline(0, color='black', lw=0.8, ls='--')
ax.set_xlabel('x')
ax.set_ylabel('f(x) = aˣ')
ax.set_title('Funciones Crecientes (a > 1)')
ax.legend(fontsize=9)
ax.annotate('Punto común (0,1)', xy=(0,1), xytext=(0.5, 2.5),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=9)

# ── Panel derecho: decrecientes ─────────────────────────────
ax = axes[1]
bases_dec = [1/1.5, 1/np.e, 0.5, 1/3, 1/5]
labels_dec = ['(1/1.5)^x', 'e^{-x}', '(0.5)^x', '(1/3)^x', '(1/5)^x']
for a, lab, col in zip(bases_dec, labels_dec, colores):
    y = a**x
    ax.plot(x, y, color=col, lw=2, label=lab)

ax.set_ylim(-0.5, 10)
ax.axhline(0, color='black', lw=0.8, ls='--')
ax.axvline(0, color='black', lw=0.8, ls='--')
ax.set_xlabel('x')
ax.set_ylabel('f(x) = aˣ')
ax.set_title('Funciones Decrecientes (0 < a < 1)')
ax.legend(fontsize=9)

plt.tight_layout()
plt.savefig('cap05_fig01_familia_exponencial.png', bbox_inches='tight', dpi=110)
plt.show()
print("✓ Figura guardada: cap05_fig01_familia_exponencial.png")

# Verificación numérica del número e
n_vals = [10, 100, 1000, 10000, 100000]
print("\n─── Convergencia al número e ───")
for n in n_vals:
    e_aprox = (1 + 1/n)**n
    error = abs(e_aprox - np.e)
    print(f"  n={n:>7,}: e ≈ {e_aprox:.8f}  |error| = {error:.2e}")
print(f"  Valor exacto: e = {np.e:.10f}")
```

## Sección 5.2 — Derivadas de Funciones Exponenciales
### Ejemplo 5.2 — Derivadas de funciones exponenciales y sus aplicaciones económicas

**Fórmulas:**
- $\frac{d}{dx}(e^x) = e^x$
- $\frac{d}{dx}(e^{u}) = e^{u} \cdot u'$ (regla de la cadena)
- $\frac{d}{dx}(a^x) = a^x \ln a$

**Aplicaciones económicas:**
- **(a)** $f(x)=3e^{2x^2}$ — costo con crecimiento acelerado  
- **(b)** $g(t)=e^{-0.05t}(2t+1)$ — modelo de descuento  
- **(c)** $h(x)=\frac{e^x}{e^x+1}$ — función sigmoide / logística  
- **(d)** $V(t)=50000 \cdot (0.85)^t$ — depreciación anual 15%  


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 5.2 — Derivadas de funciones exponenciales
# ═══════════════════════════════════════════════════════════════
x = np.linspace(-2, 3, 400)
t = np.linspace(0, 10, 400)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Ejemplo 5.2 — Derivadas de Funciones Exponenciales', fontsize=14, fontweight='bold')

# ── (a) f(x) = 3e^{2x²} ─────────────────────────────────────
ax = axes[0,0]
xa = np.linspace(-1.5, 1.5, 400)
f_a = 3*np.exp(2*xa**2)
df_a = 12*xa*np.exp(2*xa**2)  # f'(x) = 12x·e^{2x²}
ax.plot(xa, f_a, color=C_AZUL, lw=2.5, label=r'$f(x)=3e^{2x^2}$')
ax.plot(xa, df_a, color=C_ROJO, lw=2, ls='--', label=r"$f'(x)=12x\,e^{2x^2}$")
ax.axhline(0, color='black', lw=0.5)
ax.set_ylim(-20, 60)
ax.set_xlabel('x')
ax.set_title('(a) Costo con Crecimiento Acelerado')
ax.legend(fontsize=9)
# Verificación numérica
x0 = 1.0
print(f"─── Verificación (a): f'(1.0) ───")
print(f"  Analítico: f'(1) = 12·1·e^2 = {12*1*np.exp(2):.4f}")
print(f"  Numérico:  Δf/Δx ≈ {(3*np.exp(2*(1.001)**2)-3*np.exp(2*1**2))/0.001:.4f}")

# ── (b) g(t) = e^{-0.05t}(2t+1) ────────────────────────────
ax = axes[0,1]
g_b = np.exp(-0.05*t)*(2*t + 1)
dg_b = np.exp(-0.05*t)*(1.95 - 0.1*t)  # g'(t) = e^{-0.05t}(1.95 - 0.1t)
ax.plot(t, g_b, color=C_VERDE, lw=2.5, label=r'$g(t)=e^{-0.05t}(2t+1)$')
ax.plot(t, dg_b, color=C_NARANJA, lw=2, ls='--', label=r"$g'(t)=e^{-0.05t}(1.95-0.1t)$")
ax.axhline(0, color='black', lw=0.8)
# Máximo cuando 1.95 - 0.1t = 0 → t = 19.5
t_max = 19.5
ax.set_xlabel('t (tiempo)')
ax.set_title('(b) Modelo de Descuento — Regla del Producto')
ax.legend(fontsize=9)
print(f"
─── Verificación (b): g'(10.0) ───")
print(f"  Analítico: {np.exp(-0.05*10)*(1.95-0.1*10):.4f}")
print(f"  Máximo en t = {t_max} (cuando g'(t)=0)")

# ── (c) h(x) = e^x/(e^x+1) — Sigmoide ─────────────────────
ax = axes[1,0]
xc = np.linspace(-5, 5, 400)
h_c = np.exp(xc)/(np.exp(xc) + 1)
dh_c = np.exp(xc)/(np.exp(xc)+1)**2  # h'(x) = e^x/(e^x+1)²
ax.plot(xc, h_c, color=C_MORADO, lw=2.5, label=r'$h(x)=rac{e^x}{e^x+1}$ (sigmoide)')
ax.plot(xc, dh_c, color=C_CAFE, lw=2, ls='--', label=r"$h'(x)=rac{e^x}{(e^x+1)^2}$")
ax.axhline(0.5, color='gray', lw=0.8, ls=':')
ax.axhline(1.0, color='gray', lw=0.8, ls=':')
ax.axhline(0.0, color='gray', lw=0.8, ls=':')
ax.set_xlabel('x')
ax.set_title('(c) Función Sigmoide — Modelo Logístico Simplificado')
ax.legend(fontsize=9)
ax.annotate('h(0)=0.5', xy=(0,0.5), xytext=(1.5,0.4),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=9)

# ── (d) V(t) = 50000·(0.85)^t — Depreciación ───────────────
ax = axes[1,1]
td = np.linspace(0, 20, 400)
V_d = 50000*(0.85)**td
dV_d = 50000*(0.85)**td * np.log(0.85)  # V'(t) = 50000·(0.85)^t·ln(0.85)
ax.plot(td, V_d, color=C_AZUL, lw=2.5, label=r'$V(t)=50{,}000\cdot(0.85)^t$')
ax.plot(td, -dV_d, color=C_ROJO, lw=2, ls='--', label=r'$|V'(t)|$ — tasa depreciación')
ax.fill_between(td, 0, V_d, alpha=0.15, color=C_AZUL)
ax.set_xlabel('t (años)')
ax.set_ylabel('Soles')
ax.set_title('(d) Depreciación Anual al 15%')
ax.legend(fontsize=9)
# Valor residual 10%
t_vida = np.log(0.1)/np.log(0.85)
ax.axvline(t_vida, color=C_VERDE, lw=1.5, ls=':', label=f'Vida útil: {t_vida:.1f} años')
ax.text(t_vida+0.5, 5000, f't={t_vida:.1f} años
(10% valor)', fontsize=8, color=C_VERDE)
print(f"
─── Depreciación (d) ───")
print(f"  V(0)  = S/ {50000:.0f}")
print(f"  V(5)  = S/ {50000*(0.85)**5:.2f}")
print(f"  V(10) = S/ {50000*(0.85)**10:.2f}")
print(f"  V'(0) = S/ {50000*np.log(0.85):.2f} por año (tasa inicial)")
print(f"  Vida útil contable (10%): {t_vida:.1f} años")

plt.tight_layout()
plt.savefig('cap05_fig02_derivadas_exponenciales.png', bbox_inches='tight', dpi=110)
plt.show()
print("\n✓ Figura guardada: cap05_fig02_derivadas_exponenciales.png")
```

## Sección 5.3 — La Función Logarítmica Natural
### Ejemplo 5.3 — Derivadas con logaritmos

**Fórmulas:**
- $\frac{d}{dx}(\ln x) = \frac{1}{x}$
- $\frac{d}{dx}(\ln|u|) = \frac{u'}{u}$

**Casos:**  
- $f(x) = \ln(3x^2+5)$  
- $g(x) = \ln\left(\frac{x^2+1}{x-1}\right)$  
- $h(x) = x^2 \ln x$  
- $C(x) = 100 + 50\ln(1+0.1x)$ — costo con economías de escala  


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 5.3 — Función Logarítmica y sus Derivadas
# ═══════════════════════════════════════════════════════════════
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Ejemplo 5.3 — Derivadas de Funciones Logarítmicas', fontsize=14, fontweight='bold')

# ── (a) f(x) = ln(3x² + 5) ──────────────────────────────────
ax = axes[0,0]
xa = np.linspace(-3, 3, 400)
f_a = np.log(3*xa**2 + 5)
df_a = 6*xa/(3*xa**2 + 5)  # f'(x) = 6x/(3x²+5)
ax.plot(xa, f_a, color=C_AZUL, lw=2.5, label=r'$f(x)=\ln(3x^2+5)$')
ax.plot(xa, df_a, color=C_ROJO, lw=2, ls='--', label=r"$f'(x)=rac{6x}{3x^2+5}$")
ax.axhline(0, color='black', lw=0.5)
ax.set_xlabel('x')
ax.set_title('(a) Logaritmo de Función Cuadrática')
ax.legend(fontsize=9)
print("─── Verificación f'(2) ───")
print(f"  Analítico: {6*2/(3*4+5):.4f}")
print(f"  Numérico:  {(np.log(3*(2.001)**2+5)-np.log(3*4+5))/0.001:.4f}")

# ── (b) g(x) = ln((x²+1)/(x-1)) para x > 1 ─────────────────
ax = axes[0,1]
xb = np.linspace(1.05, 5, 400)
g_b = np.log((xb**2+1)/(xb-1))
dg_b = 2*xb/(xb**2+1) - 1/(xb-1)  # g'(x) = 2x/(x²+1) - 1/(x-1)
ax.plot(xb, g_b, color=C_VERDE, lw=2.5, label=r'$g(x)=\ln\left(rac{x^2+1}{x-1}ight)$')
ax.plot(xb, dg_b, color=C_NARANJA, lw=2, ls='--', label=r"$g'(x)=rac{2x}{x^2+1}-rac{1}{x-1}$")
ax.axhline(0, color='black', lw=0.8)
ax.set_xlim(1.05, 5)
ax.set_ylim(-3, 10)
ax.set_xlabel('x')
ax.set_title('(b) Logaritmo de Cociente (x > 1)')
ax.legend(fontsize=9)

# ── (c) h(x) = x² ln x ──────────────────────────────────────
ax = axes[1,0]
xc = np.linspace(0.01, 4, 400)
h_c = xc**2 * np.log(xc)
dh_c = 2*xc*np.log(xc) + xc  # h'(x) = x(2ln x + 1)
ax.plot(xc, h_c, color=C_MORADO, lw=2.5, label=r'$h(x)=x^2\ln x$')
ax.plot(xc, dh_c, color=C_CAFE, lw=2, ls='--', label=r"$h'(x)=x(2\ln x+1)$")
ax.axhline(0, color='black', lw=0.8)
# Mínimo de h: h'=0 → x = e^{-1/2}
x_min = np.exp(-0.5)
ax.axvline(x_min, color='gray', lw=1, ls=':', alpha=0.7)
ax.annotate(f'Mín en x={x_min:.3f}', xy=(x_min, x_min**2*np.log(x_min)),
            xytext=(1.5, -0.2), arrowprops=dict(arrowstyle='->', color='black'), fontsize=8)
ax.set_ylim(-0.5, 8)
ax.set_xlabel('x')
ax.set_title('(c) Producto Polinomio × Logaritmo')
ax.legend(fontsize=9)

# ── (d) C(x) = 100 + 50·ln(1+0.1x) — Costo ────────────────
ax = axes[1,1]
xd = np.linspace(0, 200, 400)
C_d = 100 + 50*np.log(1 + 0.1*xd)
CM_d = 5/(1 + 0.1*xd)  # C'(x) = 5/(1+0.1x)
C_medio = C_d/xd  # Costo medio (evitar x=0)
xd2 = np.linspace(0.1, 200, 400)
C_d2 = 100 + 50*np.log(1 + 0.1*xd2)
CM_d2 = 5/(1 + 0.1*xd2)
Cmed_d2 = C_d2/xd2
ax.plot(xd, C_d, color=C_AZUL, lw=2.5, label='C(x) — Costo Total')
ax.plot(xd, 5/(1+0.1*xd)*20, color=C_ROJO, lw=2, ls='--', label='CM(x) — Costo Marginal ×20')
ax.set_xlabel('x (unidades)')
ax.set_ylabel('Miles de Soles')
ax.set_title('(d) Costo con Economías de Escala Logarítmicas')
ax.legend(fontsize=9)
print(f"
─── Costo marginal logarítmico ───")
for xi in [0, 50, 100, 200]:
    cm = 5/(1+0.1*xi)
    print(f"  CM({xi:3d}) = {cm:.3f}  (→ decreciente: economías de escala)")

plt.tight_layout()
plt.savefig('cap05_fig03_derivadas_logaritmicas.png', bbox_inches='tight', dpi=110)
plt.show()
print("\n✓ Figura guardada: cap05_fig03_derivadas_logaritmicas.png")
```

## Sección 5.4 — Diferenciación Logarítmica
### Ejemplo 5.4 — Casos fundamentales: $y=x^x$, $y=x^{\sqrt{x}}$, función compleja

**Técnica:** Tomar $\ln y$, aplicar propiedades, derivar implícitamente.


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 5.4 — Diferenciación Logarítmica
# ═══════════════════════════════════════════════════════════════
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Ejemplo 5.4 — Diferenciación Logarítmica', fontsize=14, fontweight='bold')

# ── (a) y = x^x ─────────────────────────────────────────────
ax = axes[0]
x_a = np.linspace(0.05, 2.5, 400)
y_a = x_a**x_a                           # y = x^x
dy_a = x_a**x_a * (np.log(x_a) + 1)     # y' = x^x(ln x + 1)
ax.plot(x_a, y_a, color=C_AZUL, lw=2.5, label=r'$y = x^x$')
ax.plot(x_a, dy_a, color=C_ROJO, lw=2, ls='--', label=r"$y' = x^x(\ln x + 1)$")
ax.axhline(0, color='black', lw=0.5)
# Mínimo: y'=0 → ln x + 1 = 0 → x = 1/e
x_min_a = 1/np.e
y_min_a = x_min_a**x_min_a
ax.plot(x_min_a, y_min_a, 'go', ms=8, label=f'Mín: ({x_min_a:.3f}, {y_min_a:.3f})')
ax.set_xlabel('x')
ax.set_title(r'$y = x^x$ — Potencia Variable')
ax.legend(fontsize=9)
ax.set_ylim(-0.5, 4)
print("─── y = x^x ───")
print(f"  Mínimo en x = 1/e = {1/np.e:.4f},  y_min = (1/e)^(1/e) = {(1/np.e)**(1/np.e):.4f}")
print(f"  Verificación y'(2): {2**2*(np.log(2)+1):.4f}")

# ── (b) y = x^{√x} ──────────────────────────────────────────
ax = axes[1]
x_b = np.linspace(0.01, 4, 400)
y_b = x_b**np.sqrt(x_b)
dy_b = x_b**np.sqrt(x_b) * (np.log(x_b) + 2) / (2*np.sqrt(x_b))
ax.plot(x_b, y_b, color=C_VERDE, lw=2.5, label=r'$y = x^{\sqrt{x}}$')
ax.plot(x_b, dy_b, color=C_NARANJA, lw=2, ls='--', label=r"$y' = x^{\sqrt{x}}\cdotrac{\ln x+2}{2\sqrt{x}}$")
ax.axhline(0, color='black', lw=0.5)
ax.set_ylim(-0.5, 6)
ax.set_xlabel('x')
ax.set_title(r'$y = x^{\sqrt{x}}$ — Diferenciación Log.')
ax.legend(fontsize=9)
print("
─── y = x^√x ───")
x_test = 4
y_test = x_test**np.sqrt(x_test)
dy_test = y_test*(np.log(x_test)+2)/(2*np.sqrt(x_test))
print(f"  y(4) = 4^2 = {y_test:.4f}")
print(f"  y'(4) = {dy_test:.4f}")

# ── (c) Función compleja y = x³(2x+1)⁴/√(x²+1) ─────────────
ax = axes[2]
x_c = np.linspace(0.1, 3, 400)
y_c = x_c**3 * (2*x_c+1)**4 / np.sqrt(x_c**2+1)
# y'/y = 3/x + 8/(2x+1) - x/(x²+1)
dy_c = y_c * (3/x_c + 8/(2*x_c+1) - x_c/(x_c**2+1))
ax.plot(x_c, y_c, color=C_MORADO, lw=2.5, label=r'$y = rac{x^3(2x+1)^4}{\sqrt{x^2+1}}$')
ax.plot(x_c, dy_c, color=C_CAFE, lw=2, ls='--', label=r"$y'$ — dif. logarítmica")
ax.set_xlabel('x')
ax.set_title('(c) Función Compleja — Dif. Logarítmica')
ax.legend(fontsize=9)
ax.set_ylim(-50, 600)
print("
─── Función compleja y'(1) ───")
x1 = 1.0
y1 = 1**3*(2+1)**4/np.sqrt(2)
dy1 = y1*(3/1 + 8/3 - 1/2)
print(f"  y(1)  = 1·81/√2 = {y1:.4f}")
print(f"  y'(1) = {dy1:.4f}")

plt.tight_layout()
plt.savefig('cap05_fig04_dif_logaritmica.png', bbox_inches='tight', dpi=110)
plt.show()
print("\n✓ Figura guardada: cap05_fig04_dif_logaritmica.png")
```

## Sección 5.5 — Aplicaciones Financieras
### Ejemplo 5.5 — Capitalización Continua como Límite

**Teorema:** Con capitalización continua a tasa $r$:
$$A(t) = A_0 \cdot e^{rt}$$

**Comparación:** $A_0 = $ S/10,000, $r = 6\%$, horizonte 20 años.


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 5.5 — Capitalización Discreta vs. Continua
# ═══════════════════════════════════════════════════════════════
A0 = 10000   # Capital inicial
r  = 0.06    # Tasa anual 6%
T  = 20      # Horizonte 20 años
t  = np.linspace(0, T, 400)

# Modelos de capitalización
A_anual      = A0*(1 + r)**t
A_semestral  = A0*(1 + r/2)**(2*t)
A_mensual    = A0*(1 + r/12)**(12*t)
A_diaria     = A0*(1 + r/365)**(365*t)
A_continua   = A0*np.exp(r*t)
A_lineal     = A0 + A0*r*t  # Para comparar

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Ejemplo 5.5 — Capitalización: Discreta vs. Continua', fontsize=14, fontweight='bold')

# ── Panel izquierdo: todas las capitalizaciones ──────────────
ax = axes[0]
ax.plot(t, A_anual,     color=C_AZUL,    lw=1.8, label='Anual (n=1)')
ax.plot(t, A_semestral, color=C_VERDE,   lw=1.8, label='Semestral (n=2)')
ax.plot(t, A_mensual,   color=C_NARANJA, lw=1.8, label='Mensual (n=12)')
ax.plot(t, A_diaria,    color=C_MORADO,  lw=1.8, label='Diaria (n=365)')
ax.plot(t, A_continua,  color=C_ROJO,    lw=2.5, label='Continua ($e^{rt}$)', ls='-')
ax.fill_between(t, A_anual, A_continua, alpha=0.1, color=C_ROJO, label='Diferencia continua-anual')
ax.set_xlabel('Años')
ax.set_ylabel('Capital Acumulado (S/)')
ax.set_title(f'A₀=S/{A0:,}, r={r:.0%}, Horizonte={T} años')
ax.legend(fontsize=8)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'S/{x:,.0f}'))

# Tabla de valores a t=20
print("─── Tabla de Capitalización (t=20 años) ───")
print(f"  {'Tipo':<20} {'Monto':>15} {'Diferencia vs Continua':>25}")
print("  " + "─"*60)
for tipo, val in [('Anual', A_anual[-1]), ('Semestral', A_semestral[-1]),
                   ('Mensual', A_mensual[-1]), ('Diaria', A_diaria[-1]),
                   ('Continua', A_continua[-1])]:
    dif = A_continua[-1] - val
    print(f"  {tipo:<20} S/{val:>12,.2f}  {'+S/' if dif==0 else 'Dif: S/'}{abs(dif):>10,.2f}")

# ── Panel derecho: crecimiento para distintas tasas ──────────
ax = axes[1]
for tasa, col in zip([0.04, 0.06, 0.08, 0.10, 0.12],
                      [C_AZUL, C_VERDE, C_NARANJA, C_ROJO, C_MORADO]):
    A = A0*np.exp(tasa*t)
    # Tiempo de duplicación
    t_dup = np.log(2)/tasa
    ax.plot(t, A, color=col, lw=2, label=f'r={tasa:.0%} (dup. en {t_dup:.1f} a)')
    ax.axhline(2*A0, color='gray', lw=0.5, ls=':')

ax.text(0.3, 2*A0+200, f'Doble: S/{2*A0:,}', fontsize=8, color='gray')
ax.set_xlabel('Años')
ax.set_ylabel('Capital Acumulado (S/)')
ax.set_title('Crecimiento para Distintas Tasas Continuas')
ax.legend(fontsize=8)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'S/{x:,.0f}'))

plt.tight_layout()
plt.savefig('cap05_fig05_capitalizacion.png', bbox_inches='tight', dpi=110)
plt.show()
print("\n✓ Figura guardada: cap05_fig05_capitalizacion.png")
```

## Sección 5.7 — Valor Presente de Flujos Continuos
### Ejemplo 5.7 — Valor presente de una renta continua

$$VP = \int_0^T R(t)\,e^{-rt}\,dt = \frac{R}{r}(1-e^{-rT}) \text{ si } R \text{ es constante}$$

**Datos:** R=S/12,000/año, T=5 años, r=8% continuo.


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 5.7 — Valor Presente de Flujo Continuo
# ═══════════════════════════════════════════════════════════════
from scipy.integrate import quad

R = 12000   # Flujo anual constante (S/)
r = 0.08    # Tasa de descuento continua
T = 5       # Horizonte

# Cálculo analítico
VP_analitico = R/r * (1 - np.exp(-r*T))
print("─── Valor Presente de Renta Continua ───")
print(f"  R = S/{R:,}/año | r = {r:.0%} | T = {T} años")
print(f"  VP = {R}/{r:.2f}·(1 - e^(-{r}·{T}))")
print(f"  VP = {R/r:,.2f}·(1 - {np.exp(-r*T):.4f})")
print(f"  VP = S/{VP_analitico:,.2f}")

# Verificación numérica por cuadratura
VP_numerico, error = quad(lambda t: R*np.exp(-r*t), 0, T)
print(f"
  Verificación numérica: VP = S/{VP_numerico:,.2f} (error = {error:.2e})")

# VPN con inversión inicial
I0 = 45000
VPN = VP_analitico - I0
print(f"
  Inversión inicial: S/{I0:,}")
print(f"  VPN = S/{VP_analitico:,.2f} - S/{I0:,} = S/{VPN:,.2f} ({'✓ Rentable' if VPN>0 else '✗ No rentable'})")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Ejemplo 5.7 — Valor Presente de Flujos Continuos', fontsize=14, fontweight='bold')

# ── Panel izquierdo: flujo nominal y descontado ──────────────
ax = axes[0]
t_plot = np.linspace(0, T, 400)
flujo_nominal   = np.full_like(t_plot, R)
flujo_desctado  = R * np.exp(-r*t_plot)
ax.plot(t_plot, flujo_nominal,  color=C_AZUL, lw=2.5, label=f'Flujo nominal R={R:,} S//año')
ax.plot(t_plot, flujo_desctado, color=C_ROJO, lw=2.5, ls='--', label='Flujo descontado R·e^{-rt}')
ax.fill_between(t_plot, 0, flujo_desctado, alpha=0.25, color=C_ROJO, label=f'VP = S/{VP_analitico:,.0f}')
ax.set_xlabel('Tiempo (años)')
ax.set_ylabel('Flujo (S//año)')
ax.set_title('Flujo Nominal vs. Descontado')
ax.legend(fontsize=9)

# ── Panel derecho: sensibilidad VP a la tasa ─────────────────
ax = axes[1]
tasas_sens = np.linspace(0.01, 0.20, 400)
VP_sens = R/tasas_sens * (1 - np.exp(-tasas_sens*T))
ax.plot(tasas_sens*100, VP_sens, color=C_VERDE, lw=2.5)
ax.axhline(I0, color=C_ROJO, lw=1.5, ls='--', label=f'I₀ = S/{I0:,}')
ax.axhline(VP_analitico, color=C_AZUL, lw=1.5, ls=':', label=f'VP(8%) = S/{VP_analitico:,.0f}')
# TIR: donde VP = I0
tir_func = lambda r_: R/r_*(1-np.exp(-r_*T)) - I0
tir = fsolve(tir_func, 0.15)[0]
ax.axvline(tir*100, color=C_NARANJA, lw=1.5, ls='-.', label=f'TIR ≈ {tir:.1%}')
ax.set_xlabel('Tasa de Descuento (%)')
ax.set_ylabel('Valor Presente (S/)')
ax.set_title(f'Sensibilidad VP — R={R:,}, T={T} años')
ax.legend(fontsize=9)
print(f"
  TIR del proyecto: {tir:.2%}")

plt.tight_layout()
plt.savefig('cap05_fig07_valor_presente.png', bbox_inches='tight', dpi=110)
plt.show()
print("\n✓ Figura guardada: cap05_fig07_valor_presente.png")
```

## Sección 5.8 — Modelos de Crecimiento y Decaimiento Exponencial
### Ejemplo 5.8 — Depreciación y Difusión Publicitaria

**Modelo general:** $A(t) = A_0 e^{kt}$  
- $k > 0$: crecimiento  
- $k < 0$: decaimiento (depreciación)


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 5.8 — Depreciación y Difusión Publicitaria
# ═══════════════════════════════════════════════════════════════
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Ejemplo 5.8 — Crecimiento y Decaimiento Exponencial', fontsize=14, fontweight='bold')

# ── Panel izquierdo: depreciación de vehículo ───────────────
ax = axes[0]
V0 = 80000   # Valor inicial S/80,000
k  = -0.15   # Tasa decaimiento 15% anual
t_dep = np.linspace(0, 20, 400)
V_t = V0 * np.exp(k*t_dep)
dV_t = k * V_t  # V'(t) = k·V(t)

ax.plot(t_dep, V_t, color=C_AZUL, lw=2.5, label=f'V(t) = {V0:,}·e^({k}t)')
ax.fill_between(t_dep, 0, V_t, alpha=0.12, color=C_AZUL)
# Valor al 10% (vida útil contable)
t_10 = np.log(0.1)/k
ax.axhline(0.1*V0, color=C_ROJO, lw=1.5, ls='--', alpha=0.8, label=f'10% valor = S/{0.1*V0:,}')
ax.axvline(t_10, color=C_VERDE, lw=1.5, ls=':', alpha=0.8, label=f'Vida útil: {t_10:.1f} años')
# Puntos específicos
for ti, lbl in [(3, 't=3'), (5, 't=5'), (10, 't=10')]:
    vi = V0*np.exp(k*ti)
    ax.plot(ti, vi, 'o', color=C_NARANJA, ms=8)
    ax.annotate(f'S/{vi:,.0f}', xy=(ti, vi), xytext=(ti+0.3, vi+3000), fontsize=8)
ax.set_xlabel('Años (t)')
ax.set_ylabel('Valor del Vehículo (S/)')
ax.set_title('Depreciación Exponencial — Vehículo Empresarial')
ax.legend(fontsize=9)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'S/{x:,.0f}'))

print("─── Depreciación del Vehículo ───")
for ti in [0, 3, 5, 10]:
    vi = V0*np.exp(k*ti)
    dv = k*vi
    print(f"  t={ti:2d}: V={vi:8,.2f}  |  V'={dv:8,.2f} S//año")
print(f"  Vida útil (10% residual): t = {t_10:.2f} años")

# ── Panel derecho: difusión publicitaria ────────────────────
ax = axes[1]
S_max = 500000  # Techo de ventas
lam   = 0.3     # Velocidad de respuesta
t_pub = np.linspace(0, 15, 400)
S_t   = S_max*(1 - np.exp(-lam*t_pub))
dS_t  = S_max*lam*np.exp(-lam*t_pub)

ax.plot(t_pub, S_t/1000,  color=C_MORADO, lw=2.5, label=f'S(t) = {S_max/1000:.0f}k·(1-e^(-{lam}t))')
ax.plot(t_pub, dS_t/1000, color=C_CAFE,   lw=2,   ls='--', label=f"S'(t) — tasa de ventas")
# Punto t½
t_medio = np.log(2)/lam
ax.axhline(S_max/2/1000, color='gray', lw=1, ls=':', alpha=0.7)
ax.axvline(t_medio, color=C_ROJO, lw=1.5, ls='-.', label=f't₁/₂ = {t_medio:.2f} meses')
ax.set_xlabel('Meses (t)')
ax.set_ylabel('Ventas (miles de unidades)')
ax.set_title('Modelo de Difusión Publicitaria')
ax.legend(fontsize=9)

print(f"
─── Difusión Publicitaria (λ={lam}) ───")
print(f"  S_max = {S_max:,} unidades")
print(f"  t½ = {t_medio:.2f} meses (mitad de ventas máximas)")
for ti in [1, 3, 6, 12]:
    si = S_max*(1-np.exp(-lam*ti))
    print(f"  S({ti:2d}) = {si:8,.0f} unidades  ({si/S_max:.1%} del máximo)")

plt.tight_layout()
plt.savefig('cap05_fig08_crecimiento_decaimiento.png', bbox_inches='tight', dpi=110)
plt.show()
print("\n✓ Figura guardada: cap05_fig08_crecimiento_decaimiento.png")
```

## Sección 5.9 — Modelo Logístico de Crecimiento
### Ejemplo 5.9 — Plataforma Digital: usuarios en el tiempo

$$N(t) = \frac{K}{1 + A\,e^{-rt}}$$

**Punto de inflexión:** $t_0 = \frac{\ln A}{r}$, donde $N(t_0) = K/2$ y la tasa de crecimiento es **máxima**.


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 5.9 — Modelo Logístico de Crecimiento
# ═══════════════════════════════════════════════════════════════
K   = 500    # Capacidad máxima (miles de usuarios)
A   = 99     # Parámetro de posición: A=(K-N0)/N0
r   = 0.5    # Tasa de crecimiento
t_log = np.linspace(0, 20, 500)

# Función logística y su derivada
N_t  = K / (1 + A*np.exp(-r*t_log))
dN_t = r*N_t*(1 - N_t/K)

# Parámetros clave
N0    = K/(1+A)
t_inf = np.log(A)/r    # Punto de inflexión
N_inf = K/2
tasa_max = r*K/4       # Tasa máxima en el inflexión

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Ejemplo 5.9 — Modelo Logístico: Plataforma Digital', fontsize=14, fontweight='bold')

# ── Panel izquierdo: curva S ─────────────────────────────────
ax = axes[0]
ax.plot(t_log, N_t, color=C_AZUL, lw=3, label=f'N(t) = {K}/(1+{A}·e^(-{r}t))')
ax.axhline(K, color='gray', lw=1.5, ls='--', label=f'Capacidad K={K}k')
ax.axhline(K/2, color=C_VERDE, lw=1, ls=':', label=f'K/2={K//2}k')
ax.axvline(t_inf, color=C_ROJO, lw=1.5, ls='-.', label=f'Inflexión t₀={t_inf:.2f} meses')
ax.plot(t_inf, K/2, 'r*', ms=12, label=f'N(t₀)={K//2}k usuarios')
# Fase de crecimiento / saturación
ax.fill_between(t_log[t_log<=t_inf], 0, N_t[t_log<=t_inf], alpha=0.15, color=C_VERDE, label='Fase crecimiento')
ax.fill_between(t_log[t_log>=t_inf], 0, N_t[t_log>=t_inf], alpha=0.10, color=C_NARANJA, label='Fase saturación')
ax.set_xlabel('Meses (t)')
ax.set_ylabel('Usuarios (miles)')
ax.set_title(f'Curva S — N(0)={N0:.1f}k, K={K}k usuarios')
ax.legend(fontsize=8, loc='upper left')

print("─── Modelo Logístico — Plataforma Digital ───")
print(f"  Usuarios iniciales N(0) = {N0:.2f}k")
print(f"  Capacidad máxima K      = {K}k")
print(f"  Punto inflexión t₀      = {t_inf:.2f} meses")
print(f"  Tasa máx. N'(t₀)        = {tasa_max:.1f}k usuarios/mes")

# ── Panel derecho: tasa de crecimiento N'(t) ─────────────────
ax = axes[1]
ax.plot(t_log, dN_t, color=C_ROJO, lw=2.5, label="N'(t) = r·N·(1-N/K)")
ax.fill_between(t_log, 0, dN_t, alpha=0.2, color=C_ROJO)
ax.axvline(t_inf, color=C_AZUL, lw=1.5, ls='--', label=f'Máx en t₀={t_inf:.2f}')
ax.plot(t_inf, tasa_max, 'b*', ms=12, label=f"N'_máx = {tasa_max:.1f}k/mes")
ax.set_xlabel('Meses (t)')
ax.set_ylabel("Tasa de Crecimiento (miles usuarios/mes)")
ax.set_title("N'(t) — Tasa de Incorporación de Usuarios")
ax.legend(fontsize=9)

print("
  Evolución temporal:")
for ti in [0, 3, 9, int(t_inf), 15, 20]:
    ni = K/(1+A*np.exp(-r*ti))
    dni = r*ni*(1-ni/K)
    print(f"  t={ti:2d}: N={ni:5.1f}k  N'={dni:5.1f}k/mes  ({ni/K:.1%} capacidad)")

plt.tight_layout()
plt.savefig('cap05_fig09_logistica.png', bbox_inches='tight', dpi=110)
plt.show()
print("\n✓ Figura guardada: cap05_fig09_logistica.png")
```

## Sección 5.10 — Elasticidad con Funciones Exponenciales y Ley de Engel
### Ejemplo 5.10a — Demanda exponencial: $Q = 400e^{-0.05P}$

$$\varepsilon(P) = -0.05P \quad \Rightarrow \quad \text{Precio óptimo: } P^* = 20$$

### Ejemplo 5.10b — Ley de Engel: $Q_i = a + b\ln Y$


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 5.10 — Elasticidad y Ley de Engel
# ═══════════════════════════════════════════════════════════════
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Ejemplo 5.10 — Elasticidad Exponencial y Ley de Engel', fontsize=14, fontweight='bold')

# ── (a) Función de demanda Q = 400e^{-0.05P} ─────────────────
ax = axes[0,0]
P = np.linspace(0.5, 60, 400)
Q_P = 400*np.exp(-0.05*P)
ax.plot(P, Q_P, color=C_AZUL, lw=2.5)
ax.set_xlabel('Precio P (S/)')
ax.set_ylabel('Cantidad Q')
ax.set_title(r'Demanda: $Q = 400e^{-0.05P}$')
# Marcar zonas elástica/inelástica/unitaria
P_star = 20   # donde |ε|=1
ax.axvline(P_star, color=C_ROJO, lw=1.5, ls='--')
ax.text(5, 50, '|ε|<1
Inelástica', color=C_VERDE, fontsize=10, fontweight='bold')
ax.text(28, 100, '|ε|>1
Elástica', color=C_ROJO, fontsize=10, fontweight='bold')
ax.annotate(f'P*={P_star} (|ε|=1)', xy=(P_star, 400*np.exp(-0.05*P_star)),
            xytext=(25, 200), arrowprops=dict(arrowstyle='->', color='black'), fontsize=9)

# ── (b) Ingreso total I(P) = P·Q ─────────────────────────────
ax = axes[0,1]
IT_P = P * Q_P
ax.plot(P, IT_P, color=C_VERDE, lw=2.5, label='I(P) = P·Q')
ax.axvline(P_star, color=C_ROJO, lw=1.5, ls='--', label=f'P*={P_star} (IT máx.)')
ax.plot(P_star, P_star*400*np.exp(-0.05*P_star), 'r*', ms=12,
        label=f"IT_máx = {P_star*400*np.exp(-0.05*P_star):.1f}")
ax.set_xlabel('Precio P (S/)')
ax.set_ylabel('Ingreso Total (S/)')
ax.set_title('Ingreso Total — Máximo en |ε|=1')
ax.legend(fontsize=9)

IT_max = P_star*400*np.exp(-0.05*P_star)
print("─── Elasticidad Demanda Exponencial ───")
print(f"  Q = 400·e^(-0.05·P) → ε(P) = -0.05·P")
for p_val in [10, 20, 30]:
    eps = -0.05*p_val
    q_val = 400*np.exp(-0.05*p_val)
    it_val = p_val*q_val
    tipo = "Unitaria" if abs(eps)==1 else ("Elástica" if abs(eps)>1 else "Inelástica")
    print(f"  P={p_val:2d}: Q={q_val:.2f}  ε={eps:.2f}  IT={it_val:.2f}  → {tipo}")
print(f"
  Precio óptimo P*={P_star}  IT_máx={IT_max:.2f}")

# ── (c) Elasticidad ε(P) = -0.05P ───────────────────────────
ax = axes[1,0]
eps_P = -0.05*P
ax.plot(P, abs(eps_P), color=C_MORADO, lw=2.5, label='|ε(P)| = 0.05·P')
ax.axhline(1, color=C_ROJO, lw=1.5, ls='--', label='|ε|=1 (unitaria)')
ax.axvline(P_star, color=C_ROJO, lw=1, ls=':', alpha=0.7)
ax.fill_between(P[P<P_star], 0, abs(eps_P[P<P_star]), alpha=0.2, color=C_VERDE, label='Zona inelástica')
ax.fill_between(P[P>P_star], 0, abs(eps_P[P>P_star]), alpha=0.2, color=C_ROJO, label='Zona elástica')
ax.set_xlabel('Precio P (S/)')
ax.set_ylabel('|Elasticidad|')
ax.set_title('Elasticidad Precio de la Demanda')
ax.legend(fontsize=9)
ax.set_ylim(0, 4)

# ── (d) Ley de Engel: Q_i = a + b·ln(Y) ────────────────────
ax = axes[1,1]
Y = np.linspace(2100, 20000, 400)
# Parámetros para distintos tipos de bienes
bienes = [
    ('Bien necesario (b=30, a=-50)', -50, 30, C_AZUL, ':', True),
    ('Bien normal (b=80, a=-200)',   -200, 80, C_VERDE, '--', True),
    ('Bien de lujo (b=200, a=-600)', -600, 200, C_NARANJA, '-', True),
    ('Bien inferior (b=-20, a=300)', 300, -20, C_ROJO, '-.', True),
]
Y_umbral = 2000

for label, a, b, col, ls, show in bienes:
    Q_eng = a + b*np.log(Y)
    Q_eng_pos = np.where(Q_eng > 0, Q_eng, np.nan)
    ax.plot(Y, Q_eng_pos, color=col, lw=2, ls=ls, label=label)

ax.axhline(0, color='black', lw=0.8)
ax.set_xlabel('Ingreso Y (S/)')
ax.set_ylabel('Cantidad Demandada Q')
ax.set_title('Ley de Engel: Q = a + b·ln(Y)')
ax.legend(fontsize=8)

# Elasticidad ingreso
print("
─── Ley de Engel: Elasticidades Ingreso ───")
Y0 = 5000
for label, a, b, col, ls, show in bienes:
    Q0 = a + b*np.log(Y0)
    if Q0 > 0:
        eta = b/Q0
        tipo = 'Normal' if eta > 0 else 'Inferior'
        lujo = '(lujo)' if eta > 1 else '(necesario)' if 0 < eta < 1 else ''
        print(f"  {label[:25]}: Q(Y=5000)={Q0:.1f}  η={eta:.3f}  {tipo} {lujo}")

plt.tight_layout()
plt.savefig('cap05_fig10_elasticidad_engel.png', bbox_inches='tight', dpi=110)
plt.show()
print("\n✓ Figura guardada: cap05_fig10_elasticidad_engel.png")
```

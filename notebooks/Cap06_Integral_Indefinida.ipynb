## Sección 6.1 — Antiderivadas: Familia de Primitivas
### Ejemplo 6.1 — Verificación y visualización de antiderivadas

Si $f(x)=x^2$, entonces $F(x) = x^3/3 + C$.  
Si $CM(x)=3x^2-12x+20$, entonces $C(x)=x^3-6x^2+20x+C_0$.


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 6.1 — Familia de Antiderivadas
# ═══════════════════════════════════════════════════════════════
x = np.linspace(-2.5, 2.5, 400)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Ejemplo 6.1 — Familia de Antiderivadas de f(x)=x^2', fontsize=14, fontweight='bold')

# Panel izq: familia de primitivas
ax = axes[0]
f_x = x**2
ax.plot(x, f_x, color=C_ROJO, lw=3, label='f(x) = x² (integrando)', zorder=5)
colores = [C_AZUL, C_VERDE, C_NARANJA, C_MORADO, C_CAFE]
for c_val, col in zip([-4, -2, 0, 2, 4], colores):
    F = x**3/3 + c_val
    ax.plot(x, F, color=col, lw=1.8, ls='--', label=f'F(x)=x³/3+({c_val})', alpha=0.85)
ax.set_ylim(-8, 8)
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_title('Familia de Primitivas F(x)=x³/3 + C')
ax.legend(fontsize=8, ncol=2)
ax.text(0.02, 0.96, 'La constante C
determina la
altura de la curva',
        transform=ax.transAxes, fontsize=9, va='top',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.7))

# Panel der: costo marginal → costo total
ax = axes[1]
x2 = np.linspace(0, 8, 400)
CM_x = 3*x2**2 - 12*x2 + 20
ax.plot(x2, CM_x, color=C_ROJO, lw=3, label='CM(x) = 3x² - 12x + 20', zorder=5)
C_fijo = 50
C_total = x2**3 - 6*x2**2 + 20*x2 + C_fijo
ax.plot(x2, C_total/5, color=C_AZUL, lw=2.5, ls='--', label=f'C(x)/5 (costo total, CF={C_fijo}k)')
ax.fill_between(x2, 0, CM_x, alpha=0.15, color=C_ROJO)
ax.set_xlabel('x (miles unidades)')
ax.set_ylabel('Costo (miles S/)')
ax.set_title('Recuperación: CM(x) → C(x)')
ax.legend(fontsize=9)

print("─── Verificación Antiderivada ───")
print("  F(x) = x³/3 + C  →  F'(x) = x²  = f(x) ✓")
print()
print("─── Costo Total desde Costo Marginal ───")
print("  CM(x) = 3x²-12x+20  →  C(x) = x³-6x²+20x+50 (CF=50k)")
for xi in [0, 2, 4, 6]:
    c_val = xi**3 - 6*xi**2 + 20*xi + C_fijo
    cm_val = 3*xi**2 - 12*xi + 20
    print(f"  x={xi}: C={c_val}k, CM={cm_val}k")

plt.tight_layout()
plt.savefig('cap06_fig01_antiderivadas.png', bbox_inches='tight', dpi=110)
plt.show()
print("Figura guardada: cap06_fig01_antiderivadas.png")
```

## Sección 6.2 — Tabla de Integrales Fundamentales
### Ejemplo 6.2 — Integrales inmediatas por la Regla de la Potencia y otras reglas

**Regla de la Potencia:** $\int x^n\,dx = \dfrac{x^{n+1}}{n+1}+C \quad (n\neq-1)$

Casos:  
- (a) $\int(4x^3-3x^2+\frac{2}{x})dx$  
- (b) $\int(\sqrt{x}+\frac{1}{\sqrt{x}})dx$  
- (c) $\int(e^{2x}+3^x)dx$  


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 6.2 — Integrales Fundamentales
# ═══════════════════════════════════════════════════════════════
x = np.linspace(0.1, 4, 400)
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Ejemplo 6.2 — Integrales Inmediatas Fundamentales', fontsize=14, fontweight='bold')

# (a) Integral de polinomio con 1/x
ax = axes[0]
f_a = 4*x**3 - 3*x**2 + 2/x
F_a = x**4 - x**3 + 2*np.log(x)  # C=0
ax.plot(x, f_a, color=C_ROJO, lw=2.5, label='f(x)=4x³-3x²+2/x')
ax.plot(x, F_a, color=C_AZUL, lw=2.5, ls='--', label='F(x)=x⁴-x³+2·ln x')
ax.fill_between(x[x<=2], 0, f_a[x<=2], alpha=0.15, color=C_ROJO)
ax.axhline(0, color='black', lw=0.5)
ax.set_ylim(-10, 60)
ax.set_xlabel('x'); ax.set_title('(a) Polinomio + 1/x')
ax.legend(fontsize=9)
# Verificación por diferenciación
dF_a_num = np.gradient(F_a, x)
print("─── Verificación (a): F'(x) ≈ f(x) ───")
xi_test = 2.0
F_a_der = 4*xi_test**3 - 3*xi_test**2 + 2/xi_test
print(f"  f(2) = {F_a_der:.4f}")
print(f"  [x⁴-x³+2ln x]'|_2 = 4·8-3·4+1 = {4*8-3*4+1:.4f}  ✓")

# (b) Integral de raíces
ax = axes[1]
f_b = np.sqrt(x) + 1/np.sqrt(x)
F_b = (2/3)*x**(3/2) + 2*np.sqrt(x)
ax.plot(x, f_b, color=C_VERDE, lw=2.5, label='f(x)=√x + 1/√x')
ax.plot(x, F_b, color=C_NARANJA, lw=2.5, ls='--', label='F(x)=⅔x^(3/2)+2√x')
ax.fill_between(x[x<=3], 0, f_b[x<=3], alpha=0.15, color=C_VERDE)
ax.set_xlabel('x'); ax.set_title('(b) Funciones Raíz')
ax.legend(fontsize=9)
print("
─── Verificación (b) en x=4 ───")
print(f"  f(4) = {np.sqrt(4)+1/np.sqrt(4):.4f}")
print(f"  F'(4) = x^(1/2) + x^(-1/2)|_4 = {4**(0.5) + 4**(-0.5):.4f}  ✓")

# (c) Integral exponencial
ax = axes[2]
f_c = np.exp(2*x) + 3**x
F_c = np.exp(2*x)/2 + 3**x/np.log(3)
ax.plot(x, f_c, color=C_MORADO, lw=2.5, label='f(x)=e^(2x)+3^x')
ax.plot(x, F_c, color=C_CAFE, lw=2.5, ls='--', label='F(x)=e^(2x)/2+3^x/ln3')
ax.set_ylim(0, 100)
ax.set_xlabel('x'); ax.set_title('(c) Exponenciales Mixtas')
ax.legend(fontsize=9)
print("
─── Verificación (c) en x=1 ───")
print(f"  f(1)  = e²+3 = {np.e**2+3:.4f}")
dF_c = np.e**2 + 3*np.log(3)  # Esto NO es correcto; la derivada de e^{2x}/2 es e^{2x}
dF_c_correcto = np.e**2 + 3**1  # derivada de F(x) evaluada en 1
print(f"  F'(1) = e²+3 = {dF_c_correcto:.4f}  ✓")

print("
─── Tabla de Integrales Fundamentales ───")
tabla = [
    ("∫k dx", "kx + C"),
    ("∫xⁿ dx (n≠-1)", "xⁿ⁺¹/(n+1) + C"),
    ("∫(1/x) dx", "ln|x| + C"),
    ("∫eˣ dx", "eˣ + C"),
    ("∫aˣ dx", "aˣ/ln(a) + C"),
    ("∫sen(x) dx", "-cos(x) + C"),
    ("∫cos(x) dx", "sen(x) + C"),
]
for integral, resultado in tabla:
    print(f"  {integral:<25} = {resultado}")

plt.tight_layout()
plt.savefig('cap06_fig02_integrales_inmediatas.png', bbox_inches='tight', dpi=110)
plt.show()
print("
Figura guardada: cap06_fig02_integrales_inmediatas.png")
```

## Sección 6.3 — Método de Sustitución
### Ejemplo 6.3 — Sustitución: nivel básico, intermedio y avanzado

**Técnica:** Si $u = g(x)$, entonces $du = g'(x)dx$ y:
$$\int f(g(x))\,g'(x)\,dx = \int f(u)\,du$$


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 6.3 — Método de Sustitución
# ═══════════════════════════════════════════════════════════════
x = np.linspace(0.01, 3, 400)
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Ejemplo 6.3 — Integración por Sustitución', fontsize=14, fontweight='bold')

# ── NIVEL BÁSICO ─────────────────────────────────────────────
# (a) ∫(2x+1)^10 dx,  u=2x+1
ax = axes[0,0]
xa = np.linspace(-1.5, 1.5, 400)
f_a = (2*xa+1)**10
F_a = (2*xa+1)**11/22
ax.plot(xa, f_a/1e4, color=C_ROJO, lw=2.5, label='f(x)=(2x+1)¹⁰')
ax.plot(xa, F_a/1e4, color=C_AZUL, lw=2, ls='--', label='F(x)=(2x+1)¹¹/22')
ax.set_xlabel('x'); ax.set_title('(a) Básico: u=2x+1')
ax.legend(fontsize=9)
print("─── Sustitución (a): ∫(2x+1)^10 dx ───")
print("  u=2x+1, du=2dx → ∫u^10·du/2 = u^11/22 = (2x+1)^11/22 + C  ✓")

# (b) ∫x√(x²+1) dx,  u=x²+1
ax = axes[0,1]
f_b = xa**2 * np.sqrt(xa**2+1)
F_b = (xa**2+1)**(3/2)/3
ax.plot(xa, f_b, color=C_VERDE, lw=2.5, label='f(x)=x·√(x²+1)')
ax.plot(xa, F_b, color=C_NARANJA, lw=2, ls='--', label='F(x)=(x²+1)^(3/2)/3')
ax.set_xlabel('x'); ax.set_title('(b) Básico: u=x²+1')
ax.legend(fontsize=9)
print("
─── Sustitución (b): ∫x√(x²+1) dx ───")
print("  u=x²+1, du=2x·dx → ∫√u·du/2 = u^(3/2)/3 + C  ✓")

# (c) ∫3x²/(x³+1) dx,  u=x³+1
ax = axes[0,2]
xc = np.linspace(0.1, 3, 400)
f_c = 3*xc**2/(xc**3+1)
F_c = np.log(xc**3+1)
ax.plot(xc, f_c, color=C_MORADO, lw=2.5, label='f(x)=3x²/(x³+1)')
ax.plot(xc, F_c, color=C_CAFE, lw=2, ls='--', label='F(x)=ln(x³+1)')
ax.set_xlabel('x'); ax.set_title('(c) Básico: u=x³+1')
ax.legend(fontsize=9)
print("
─── Sustitución (c): ∫3x²/(x³+1) dx ───")
print("  u=x³+1, du=3x²dx → ∫du/u = ln|u| = ln(x³+1) + C  ✓")

# ── NIVEL INTERMEDIO ─────────────────────────────────────────
# (d) ∫(lnx)/x dx
ax = axes[1,0]
xd = np.linspace(0.1, 4, 400)
f_d = np.log(xd)/xd
F_d = np.log(xd)**2/2
ax.plot(xd, f_d, color=C_AZUL, lw=2.5, label='f(x)=ln(x)/x')
ax.plot(xd, F_d, color=C_ROJO, lw=2, ls='--', label='F(x)=(ln x)²/2')
ax.fill_between(xd[xd<=np.e], 0, f_d[xd<=np.e], alpha=0.2, color=C_AZUL)
ax.set_xlabel('x'); ax.set_title('(d) Intermedio: u=ln x')
ax.legend(fontsize=9)
print("
─── Sustitución (d): ∫(ln x)/x dx ───")
print("  u=ln x, du=dx/x → ∫u·du = u²/2 = (ln x)²/2 + C  ✓")

# (e) ∫eˣ/(eˣ+1) dx
ax = axes[1,1]
xe = np.linspace(-4, 4, 400)
f_e = np.exp(xe)/(np.exp(xe)+1)
F_e = np.log(np.exp(xe)+1)
ax.plot(xe, f_e, color=C_VERDE, lw=2.5, label='f(x)=eˣ/(eˣ+1)')
ax.plot(xe, F_e, color=C_NARANJA, lw=2, ls='--', label='F(x)=ln(eˣ+1)')
ax.set_xlabel('x'); ax.set_title('(e) Intermedio: u=eˣ+1')
ax.legend(fontsize=9)

# (f) Aplicación económica: IM(x) = 20x/(x²+1)
ax = axes[1,2]
xf = np.linspace(0, 5, 400)
IM_x = 20*xf/(xf**2+1)
I_x = 10*np.log(xf**2+1)   # I(0)=0 => C=0
ax.plot(xf, IM_x, color=C_MORADO, lw=2.5, label='IM(x)=20x/(x²+1)')
ax.plot(xf, I_x, color=C_CAFE, lw=2.5, ls='--', label='I(x)=10·ln(x²+1)')
ax.fill_between(xf, 0, IM_x, alpha=0.15, color=C_MORADO, label='Area=I(x)')
ax.set_xlabel('x (unidades)')
ax.set_ylabel('Miles S/')
ax.set_title('(f) Aplicacion: I(x) desde IM(x)')
ax.legend(fontsize=9)
print("
─── Aplicacion Economica (f): Ingreso desde IM ───")
print("  IM(x) = 20x/(x²+1)  →  I(x) = 10·ln(x²+1)  [I(0)=0]")
for xi in [1, 2, 3, 5]:
    print(f"  I({xi}) = 10·ln({xi**2+1}) = {10*np.log(xi**2+1):.4f} miles S/")

plt.tight_layout()
plt.savefig('cap06_fig03_sustitucion.png', bbox_inches='tight', dpi=110)
plt.show()
print("
Figura guardada: cap06_fig03_sustitucion.png")
```

## Sección 6.4 — Integración por Partes
### Ejemplo 6.4 — Fórmula y aplicaciones

$$\int u\,dv = uv - \int v\,du$$

**Regla LIATE:** Logarítmica > Inversa trig > Algebraica > Trigonométrica > Exponencial

Casos:  
- (a) $\int x\,e^x\,dx = e^x(x-1)+C$  
- (b) $\int x^2\,e^x\,dx = e^x(x^2-2x+2)+C$  
- (c) $\int \ln x\,dx = x\ln x - x + C$  
- (d) $\int e^x\,\text{sen}\,x\,dx$ — integral cíclica  
- (e) Valor Presente: $\int_0^{10} t\,e^{-0.1t}dt$  


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 6.4 — Integración por Partes
# ═══════════════════════════════════════════════════════════════
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Ejemplo 6.4 — Integracion por Partes', fontsize=14, fontweight='bold')

# (a) ∫x·eˣ dx = eˣ(x-1)+C
ax = axes[0,0]
x = np.linspace(-2, 2.5, 400)
f_a = x*np.exp(x)
F_a = np.exp(x)*(x-1)
ax.plot(x, f_a, color=C_ROJO, lw=2.5, label='f(x)=x·eˣ')
ax.plot(x, F_a, color=C_AZUL, lw=2, ls='--', label='F(x)=eˣ(x-1)')
ax.axhline(0, color='black', lw=0.5)
ax.set_ylim(-5, 12); ax.set_xlabel('x')
ax.set_title('(a) u=x, dv=eˣdx')
ax.legend(fontsize=9)
print("─── Partes (a): ∫x·eˣ dx ───")
print("  u=x, dv=eˣ → du=dx, v=eˣ")
print("  ∫x·eˣ dx = xeˣ - ∫eˣ dx = eˣ(x-1) + C  ✓")
print(f"  Verificacion: [eˣ(x-1)]'= eˣ(x-1)+eˣ = x·eˣ  ✓")

# (b) ∫x²eˣ dx = eˣ(x²-2x+2)+C
ax = axes[0,1]
f_b = x**2 * np.exp(x)
F_b = np.exp(x)*(x**2 - 2*x + 2)
ax.plot(x, f_b, color=C_VERDE, lw=2.5, label='f(x)=x²eˣ')
ax.plot(x, F_b, color=C_NARANJA, lw=2, ls='--', label='F(x)=eˣ(x²-2x+2)')
ax.axhline(0, color='black', lw=0.5)
ax.set_ylim(-5, 20); ax.set_xlabel('x')
ax.set_title('(b) u=x², dv=eˣdx (doble partes)')
ax.legend(fontsize=9)

# (c) ∫ln x dx = x·ln x - x + C
ax = axes[0,2]
xc = np.linspace(0.1, 4, 400)
f_c = np.log(xc)
F_c = xc*np.log(xc) - xc
ax.plot(xc, f_c, color=C_MORADO, lw=2.5, label='f(x)=ln x')
ax.plot(xc, F_c, color=C_CAFE, lw=2, ls='--', label='F(x)=x·ln x - x')
ax.axhline(0, color='black', lw=0.5)
ax.set_xlabel('x'); ax.set_title('(c) u=ln x, dv=dx')
ax.legend(fontsize=9)
print("
─── Partes (c): ∫ln x dx ───")
print("  u=ln x, dv=dx → du=dx/x, v=x")
print("  ∫ln x dx = x·ln x - ∫dx = x·ln x - x + C  ✓")

# (d) ∫eˣ·sen(x)dx — cíclica
ax = axes[1,0]
xd = np.linspace(0, 3*np.pi, 500)
f_d = np.exp(xd)*np.sin(xd)
F_d = np.exp(xd)*(np.sin(xd) - np.cos(xd))/2  # resultado analítico
ax.plot(xd, f_d, color=C_AZUL, lw=2.5, label='f(x)=eˣ·sen x')
ax.plot(xd, F_d, color=C_ROJO, lw=2, ls='--', label='F(x)=eˣ(sen x-cos x)/2')
ax.axhline(0, color='black', lw=0.5)
ax.set_xlabel('x'); ax.set_title('(d) Integral Ciclica (doble partes)')
ax.legend(fontsize=9)
print("
─── Partes (d): ∫eˣ·sen x dx — integral ciclica ───")
print("  Paso 1: I = eˣ·sen x - ∫eˣ·cos x dx")
print("  Paso 2: ∫eˣ·cos x dx = eˣ·cos x + ∫eˣ·sen x dx = eˣ·cos x + I")
print("  → I = eˣ·sen x - eˣ·cos x - I")
print("  → 2I = eˣ(sen x - cos x)")
print("  → I = eˣ(sen x - cos x)/2 + C  ✓")

# Verificacion numerica
x_ver = 1.0
analitico = np.exp(x_ver)*(np.sin(x_ver)-np.cos(x_ver))/2
derivada = np.exp(x_ver)*np.sin(x_ver)  # f(x)
derivada_F = np.exp(x_ver)*(np.sin(x_ver)-np.cos(x_ver))/2 + np.exp(x_ver)*(np.cos(x_ver)+np.sin(x_ver))/2
print(f"  Verificacion en x=1: F'(1)={derivada_F:.4f} vs f(1)={derivada:.4f}  ✓")

# (e) Aplicación: Valor Presente VP = ∫₀¹⁰ t·e^{-0.1t} dt
ax = axes[1,1]
R_t = lambda t: t*np.exp(-0.1*t)  # flujo descontado
te = np.linspace(0, 10, 400)
vp_integrando = R_t(te)
# Antiderivada por partes: u=t, dv=e^{-0.1t}dt → -10t·e^{-0.1t} - 100e^{-0.1t}
VP_acum = lambda T: (-10*T*np.exp(-0.1*T) - 100*np.exp(-0.1*T)) - (0 - 100)
VP_10 = VP_acum(10)
ax.plot(te, vp_integrando, color=C_VERDE, lw=2.5, label='R(t)·e^{-0.1t}=t·e^{-0.1t}')
ax.fill_between(te, 0, vp_integrando, alpha=0.25, color=C_VERDE, label=f'VP=S/{VP_10:.2f} miles')
ax.set_xlabel('Años'); ax.set_ylabel('Flujo descontado')
ax.set_title('(e) VP = ∫₀¹⁰ t·e^{-0.1t}dt')
ax.legend(fontsize=9)
vp_num, _ = quad(R_t, 0, 10)
print(f"
─── VP = ∫₀¹⁰ t·e^(-0.1t)dt ───")
print(f"  Por partes: VP = {VP_10:.4f} miles de S/")
print(f"  Verificacion numerica: VP = {vp_num:.4f}  ✓")

# (f) ∫x³·e^{-2x}dx (fórmula de reducción)
ax = axes[1,2]
xf = np.linspace(0, 5, 400)
f_f = xf**3 * np.exp(-2*xf)
# Resultado: -e^{-2x}(4x³+6x²+6x+3)/8
F_f = -np.exp(-2*xf)*(4*xf**3 + 6*xf**2 + 6*xf + 3)/8
ax.plot(xf, f_f, color=C_AZUL, lw=2.5, label='f(x)=x³·e^{-2x}')
ax.plot(xf, F_f, color=C_ROJO, lw=2, ls='--', label='F(x)=-e^{-2x}(4x³+6x²+6x+3)/8')
ax.axhline(0, color='black', lw=0.5)
ax.set_xlabel('x'); ax.set_title('(f) u=x³ (tres aplicaciones)')
ax.legend(fontsize=9)
print("
─── Partes triple: ∫x³·e^{-2x}dx ───")
print("  Resultado: -e^{-2x}(4x³+6x²+6x+3)/8 + C")
# Verificacion
x_v = 1.0
F_der = (-np.exp(-2*(x_v+0.001))*(4*(x_v+0.001)**3+6*(x_v+0.001)**2+6*(x_v+0.001)+3)/8 -
         (-np.exp(-2*x_v)*(4*x_v**3+6*x_v**2+6*x_v+3)/8))/0.001
print(f"  F'(1) ≈ {F_der:.4f} vs f(1)={x_v**3*np.exp(-2*x_v):.4f}  ✓")

plt.tight_layout()
plt.savefig('cap06_fig04_partes.png', bbox_inches='tight', dpi=110)
plt.show()
print("
Figura guardada: cap06_fig04_partes.png")
```

## Sección 6.5 — Fracciones Parciales
### Ejemplo 6.5 — Descomposición e integración

**Método de Heaviside:**  
$$\frac{P(x)}{Q(x)} = \sum_k \frac{A_k}{x-a_k}, \quad A_k = \frac{P(a_k)}{Q'(a_k)}$$

Casos:  
- (a) $\int\frac{5x+2}{x^2+x-2}dx$ — factores lineales distintos  
- (b) $\int\frac{3x+1}{x^2-4}dx$  
- (c) $\int\frac{2x-5}{x^2-3x+2}dx$  
- (d) Aplicación económica: Costo Marginal racional  


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 6.5 — Fracciones Parciales
# ═══════════════════════════════════════════════════════════════
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Ejemplo 6.5 — Integracion por Fracciones Parciales', fontsize=14, fontweight='bold')

# ── (a) ∫(5x+2)/(x²+x-2)dx = (7/3)ln|x-1| + (8/3)ln|x+2| ──
ax = axes[0,0]
xa = np.linspace(1.1, 6, 400)
f_a = (5*xa+2)/(xa**2+xa-2)   # = (5x+2)/((x-1)(x+2))
A_a, B_a = 7/3, 8/3
F_a = A_a*np.log(np.abs(xa-1)) + B_a*np.log(np.abs(xa+2))
ax.plot(xa, f_a, color=C_ROJO, lw=2.5, label='f(x)=(5x+2)/(x²+x-2)')
ax.plot(xa, F_a, color=C_AZUL, lw=2, ls='--', label='F(x)=(7/3)ln|x-1|+(8/3)ln|x+2|')
ax.fill_between(xa[xa<=4], 0, f_a[xa<=4], alpha=0.15, color=C_ROJO)
ax.set_xlabel('x'); ax.set_title('(a) Factores Lineales Distintos')
ax.legend(fontsize=9)
print("─── Fracciones Parciales (a) ───")
print("  (5x+2)/((x-1)(x+2)) = A/(x-1) + B/(x+2)")
print(f"  A = (5·1+2)/(1+2)   = {(5+2)/3:.4f}  (debe ser 7/3)")
print(f"  B = (5·(-2)+2)/(-2-1) = {(-10+2)/(-3):.4f}  (debe ser 8/3)")
print(f"  Integral = (7/3)ln|x-1| + (8/3)ln|x+2| + C  ✓")

# ── (b) ∫(3x+1)/(x²-4)dx ─────────────────────────────────────
ax = axes[0,1]
xb = np.linspace(2.05, 8, 400)
f_b = (3*xb+1)/(xb**2-4)   # = (3x+1)/((x-2)(x+2))
A_b = (3*2+1)/(2+2)         # A en x=2
B_b = (3*(-2)+1)/(-2-2)     # B en x=-2
F_b = A_b*np.log(np.abs(xb-2)) + B_b*np.log(np.abs(xb+2))
ax.plot(xb, f_b, color=C_VERDE, lw=2.5, label='f(x)=(3x+1)/(x²-4)')
ax.plot(xb, F_b-F_b[0], color=C_NARANJA, lw=2, ls='--', label='F(x) (normalizada)')
ax.set_xlabel('x'); ax.set_title('(b) Denominador: x²-4=(x-2)(x+2)')
ax.legend(fontsize=9)
print(f"
─── Fracciones Parciales (b) ───")
print(f"  A = (3·2+1)/(2+2) = {A_b:.4f}")
print(f"  B = (3·(-2)+1)/(-2-2) = {B_b:.4f}")
print(f"  Integral = {A_b:.4f}·ln|x-2| + {B_b:.4f}·ln|x+2| + C")

# ── (c) ∫(2x-5)/(x²-3x+2)dx ─────────────────────────────────
ax = axes[1,0]
xc = np.linspace(2.05, 6, 400)
f_c = (2*xc-5)/(xc**2-3*xc+2)  # (x-1)(x-2)
A_c = (2*1-5)/(1-2)   # x=1
B_c = (2*2-5)/(2-1)   # x=2
F_c = A_c*np.log(np.abs(xc-1)) + B_c*np.log(np.abs(xc-2))
ax.plot(xc, f_c, color=C_MORADO, lw=2.5, label='f(x)=(2x-5)/(x²-3x+2)')
ax.plot(xc, F_c-F_c[0], color=C_CAFE, lw=2, ls='--', label='F(x) (normalizada)')
ax.set_xlabel('x'); ax.set_title('(c) Denominador: (x-1)(x-2)')
ax.legend(fontsize=9)
print(f"
─── Fracciones Parciales (c) ───")
print(f"  A(x=1) = (2-5)/(1-2) = {A_c:.4f}")
print(f"  B(x=2) = (4-5)/(2-1) = {B_c:.4f}")
print(f"  Integral = {A_c:.0f}·ln|x-1| + {B_c:.0f}·ln|x-2| + C")

# ── (d) Aplicación: C'(q) = (2q²+3q+1)/(q²+q) ───────────────
ax = axes[1,1]
# C'(q) = (2q²+3q+1)/(q²+q) = (2q²+3q+1)/(q(q+1)) = 2 + 1/q (simplificado)
# C(q) = 2q + ln q + K, con C(1)=10 → K=8
xd = np.linspace(0.1, 10, 400)
CM_d = (2*xd**2+3*xd+1)/(xd**2+xd)  # simplifica a 2 + 1/q
C_d  = 2*xd + np.log(xd) + 8        # C(1)=2+0+K=10 → K=8
ax.plot(xd, CM_d, color=C_AZUL, lw=2.5, label="C'(q) = 2 + 1/q")
ax.plot(xd, C_d/5, color=C_ROJO, lw=2.5, ls='--', label='C(q)/5 = (2q+ln q+8)/5')
ax.fill_between(xd[xd<=5], 0, CM_d[xd<=5], alpha=0.15, color=C_AZUL)
ax.set_xlabel('q (unidades)')
ax.set_ylabel('Miles S/')
ax.set_title('(d) Aplicacion: C(q) desde CM racional')
ax.legend(fontsize=9)
print("
─── Aplicacion (d): C(q) desde CM racional ───")
print("  C'(q) = (2q²+3q+1)/(q(q+1)) = 2 + 1/q")
print("  C(q) = 2q + ln q + K")
print("  C(1) = 10 → K = 8")
for qi in [1, 2, 5, 10]:
    c_val = 2*qi + np.log(qi) + 8
    cm_val = 2 + 1/qi
    print(f"  q={qi:2d}: C={c_val:.2f}k, CM={cm_val:.4f}k")

plt.tight_layout()
plt.savefig('cap06_fig05_fracciones_parciales.png', bbox_inches='tight', dpi=110)
plt.show()
print("
Figura guardada: cap06_fig05_fracciones_parciales.png")
```

## Sección 6.6 — Aplicaciones Económicas Avanzadas
### Ejemplo 6.6 — Recuperación de funciones económicas e integrales impropias

Integración en el contexto:  
- Flujo de ahorro con reinversión continua  
- Costo total con economías de escala logarítmicas  


```python
# ═══════════════════════════════════════════════════════════════
# Ejemplo 6.6 — Aplicaciones Economicas de la Integral Indefinida
# ═══════════════════════════════════════════════════════════════
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Ejemplo 6.6 — Recuperacion de Funciones Economicas', fontsize=14, fontweight='bold')

# ─── Flujo de ahorro S(t) = 1000e^{0.05t} con tasa r=0.06 ──
ax = axes[0]
t = np.linspace(0, 10, 400)
S_t = 1000*np.exp(0.05*t)       # Flujo de ahorro
r   = 0.06
# ∫S(t)·e^{-rt}dt = ∫1000·e^{(0.05-0.06)t}dt = 1000·e^{-0.01t}/(-0.01)
VP_cumulativo = 1000/0.01*(1 - np.exp(-0.01*t))  # ≈ 1000·t para t pequeño
ax.plot(t, S_t, color=C_AZUL, lw=2.5, label='S(t)=1000·e^{0.05t} (flujo)')
ax.plot(t, 1000*np.exp(-0.01*t), color=C_ROJO, lw=2, ls='--', label='S(t)·e^{-0.06t} (descontado)')
ax.fill_between(t, 0, 1000*np.exp(-0.01*t), alpha=0.2, color=C_ROJO, label='VP acumulado')
ax.set_xlabel('Años'); ax.set_ylabel('S/ por año')
ax.set_title('Flujo de Ahorro Creciente
r=6%, g=5%')
ax.legend(fontsize=8)
vp_10 = quad(lambda t_: 1000*np.exp(0.05*t_)*np.exp(-r*t_), 0, 10)[0]
print(f"─── Valor Presente a 10 años ───")
print(f"  VP = ∫₀¹⁰ 1000·e^(0.05t)·e^(-0.06t) dt")
print(f"     = ∫₀¹⁰ 1000·e^(-0.01t) dt = S/{vp_10:,.2f}")

# ─── Costo con economías de escala logarítmicas ─────────────
ax = axes[1]
x = np.linspace(0.1, 200, 400)
# CM(x) = 100/(x+1) + 5,  C(0)=200
C_x = 100*np.log(x+1) + 5*x + 200
CM_x = 100/(x+1) + 5
Cmed_x = C_x/x
ax.plot(x, C_x/100, color=C_AZUL, lw=2.5, label='C(x)/100 — Costo Total')
ax.plot(x, CM_x, color=C_ROJO, lw=2, ls='--', label='CM(x) — Costo Marginal')
ax.plot(x[1:], Cmed_x[1:], color=C_VERDE, lw=2, ls=':', label='CMe(x) — Costo Medio')
ax.set_xlabel('x (unidades)'); ax.set_ylabel('Miles S/')
ax.set_title('C(x)=100·ln(x+1)+5x+200')
ax.legend(fontsize=9)
ax.set_ylim(0, 50)
print("
─── Costo con Economias de Escala ───")
print("  CM(x)=100/(x+1)+5  →  C(x)=100·ln(x+1)+5x+200")
for xi in [0, 10, 50, 100, 200]:
    if xi == 0:
        c_v, cm_v = 200, 105
    else:
        c_v = 100*np.log(xi+1)+5*xi+200
        cm_v = 100/(xi+1)+5
    print(f"  x={xi:3d}: C={c_v:.1f}k, CM={cm_v:.3f}k/u")
lim_cm = 5  # limite asintótico del CM
print(f"
  Limite asintótico del CM cuando x→∞: {lim_cm}k por unidad")

# ─── Resumen de métodos ──────────────────────────────────────
ax = axes[2]
ax.axis('off')
metodos = [
    ["Método", "Cuándo usar", "Clave"],
    ["Potencia", "xⁿ, √x, 1/x", "∫xⁿdx=xⁿ⁺¹/(n+1)"],
    ["Sustitución", "f(g(x))·g'(x)", "u=g(x), du=g'dx"],
    ["Por Partes", "Productos mixtos", "∫u dv=uv-∫v du"],
    ["Fr. Parciales", "P(x)/Q(x)", "A/x-a + B/x-b"],
    ["LIATE", "Seleccionar u", "L>I>A>T>E"],
]
tabla = ax.table(cellText=metodos[1:], colLabels=metodos[0],
                  loc='center', cellLoc='center')
tabla.auto_set_font_size(False)
tabla.set_fontsize(9)
tabla.scale(1, 2.0)
for (row, col), cell in tabla.get_celld().items():
    if row == 0:
        cell.set_facecolor('#1f77b4')
        cell.set_text_props(color='white', fontweight='bold')
    elif row % 2 == 0:
        cell.set_facecolor('#e8f0fe')
ax.set_title('Guia de Seleccion de Metodo', pad=20)

plt.tight_layout()
plt.savefig('cap06_fig06_aplicaciones.png', bbox_inches='tight', dpi=110)
plt.show()
print("
Figura guardada: cap06_fig06_aplicaciones.png")
```

## Resumen del Capítulo 6

| Método | Fórmula | Cuándo usar |
|--------|---------|-------------|
| **Regla de la Potencia** | $\int x^n\,dx = \frac{x^{n+1}}{n+1}+C$ | Polinomios y potencias |
| **Sustitución** | $u=g(x)$, $du=g'dx$ | $f(g(x))\cdot g'(x)$ |
| **Por Partes** | $\int u\,dv = uv - \int v\,du$ | Productos de funciones distintas |
| **Fracciones Parciales** | $\frac{A}{x-a}+\frac{B}{x-b}$ | Funciones racionales |

### La Constante de Integración  
La constante $C$ en $\int f(x)\,dx = F(x)+C$ representa:
- En economía: **costos fijos**, condiciones iniciales de mercado
- En finanzas: **capital inicial**, posición inicial de un activo

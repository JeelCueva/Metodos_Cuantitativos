## Ejemplo 3.1 — Tasa de Cambio Promedio e Instantánea
### $C(x)=0.5x^2+10x+200$ — Secantes que convergen a la tangente

```python
# EJEMPLO 3.1 — Tasa de cambio promedio vs instantánea
x = sp.Symbol('x')
C = 0.5*x**2 + 10*x + 200
x0 = 10

# Tasas promedio con distintos h
print('TASAS DE CAMBIO PROMEDIO en x₀=10')
print(f'{"h":>8} {"Tasa promedio":>16} {"Costo marginal":>16}')
print('-'*45)
CM_x0 = float(sp.diff(C, x).subs(x, x0))
for h in [5, 1, 0.5, 0.1, 0.01, 0.001]:
    tasa = float((C.subs(x, x0+h) - C.subs(x, x0)) / h)
    print(f'{h:>8.3f} {tasa:>16.6f} {CM_x0:>16.6f}')

print(f'\nCosto marginal exacto C\'(10) = {CM_x0}')

# Gráfica: secantes convergiendo a tangente
x_vals = np.linspace(0, 25, 300)
C_vals = 0.5*x_vals**2 + 10*x_vals + 200
C_x0 = 0.5*x0**2 + 10*x0 + 200

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
ax = axes[0]
ax.plot(x_vals, C_vals, color=C_AZUL, lw=2.5, label='$C(x)=0.5x^2+10x+200$')
colores_h = [C_ROJO, C_NARANJA, C_VERDE, C_MORADO]
for h_v, col in zip([5, 2, 1, 0.1], colores_h):
    C_xh = 0.5*(x0+h_v)**2 + 10*(x0+h_v) + 200
    m_sec = (C_xh - C_x0) / h_v
    x_sec = np.array([x0-2, x0+h_v+2])
    y_sec = C_x0 + m_sec*(x_sec - x0)
    ax.plot(x_sec, y_sec, color=col, lw=1.5, ls='--', label=f'Secante h={h_v}')
# Tangente
x_tan = np.array([x0-5, x0+7])
y_tan = C_x0 + CM_x0*(x_tan - x0)
ax.plot(x_tan, y_tan, 'k-', lw=2.5, label=f'Tangente (m={CM_x0})')
ax.scatter([x0],[C_x0],color='black',s=100,zorder=6)
ax.set_xlabel('x'); ax.set_ylabel('C(x)')
ax.set_title('Secantes → Tangente cuando h→0'); ax.legend(fontsize=8)

# Panel 2: función de ingreso con ingreso marginal
ax2 = axes[1]
I_vals = 5*x_vals + 0.5*x_vals**2
IM_vals = 5 + x_vals  # derivada
ax2.plot(x_vals, I_vals, color=C_VERDE, lw=2.5, label='$I(x)=5x+0.5x^2$')
ax2_twin = ax2.twinx()
ax2_twin.plot(x_vals, IM_vals, color=C_ROJO, lw=2, ls='--', label="$I'(x)=5+x$")
ax2_twin.set_ylabel("Ingreso marginal $I'(x)$", color=C_ROJO)
ax2.set_xlabel('Helados vendidos x'); ax2.set_ylabel('Ingreso $I(x)$', color=C_VERDE)
ax2.set_title('Ingreso Total e Ingreso Marginal')
lines1,labels1 = ax2.get_legend_handles_labels()
lines2,labels2 = ax2_twin.get_legend_handles_labels()
ax2.legend(lines1+lines2, labels1+labels2, fontsize=9)
plt.tight_layout(); plt.savefig('fig_c03_secante_tangente.png', bbox_inches='tight')
plt.show()
```

## Ejemplo 3.2 — Reglas Básicas de Derivación
### Potencia, exponencial, logaritmo — aplicaciones económicas

```python
# EJEMPLO 3.2 — Reglas básicas de derivación
x = sp.Symbol('x', positive=True)
t = sp.Symbol('t', positive=True)

# (a) Ingreso total: I(x) = 80x - 0.5x²
I = 80*x - 0.5*x**2
IM = sp.diff(I, x)
x_max_I = sp.solve(IM, x)[0]
print(f'I(x) = 80x - 0.5x²  →  IM(x) = {IM}')
print(f'IM = 0 en x = {x_max_I}  → Ingreso máximo')

# (b) Costo total: C(x) = 2x³-15x²+36x+100
C = 2*x**3 - 15*x**2 + 36*x + 100
CM = sp.diff(C, x)
print(f'C(x) = 2x³-15x²+36x+100  →  C\'(x) = {CM}')

# (c) Producción con raíz: Q(L) = 50√L
L = sp.Symbol('L', positive=True)
Q = 50*sp.sqrt(L)
PMgL = sp.diff(Q, L)
print(f'Q(L) = 50√L  →  PMgL = {PMgL}  (decreciente → RMD)')

# (d) Depreciación: V(t) = 45000 e^(-0.15t)
V = 45000*sp.exp(-0.15*t)
dV = sp.diff(V, t)
print(f'V(t) = 45000e^(-0.15t)  →  V\'(t) = {sp.simplify(dV)}')
print(f'V\'(0) = {float(dV.subs(t,0)):,.0f} S/año (tasa inicial depreciación)')

# Gráficas
x_v = np.linspace(0, 90, 400)
I_v  = 80*x_v - 0.5*x_v**2
IM_v = 80 - x_v
L_v  = np.linspace(0.1, 100, 400)
Q_v  = 50*np.sqrt(L_v)
PMg_v= 25/np.sqrt(L_v)
t_v  = np.linspace(0, 25, 400)
V_v  = 45000*np.exp(-0.15*t_v)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
ax=axes[0]
ax.plot(x_v, I_v, color=C_VERDE, lw=2.5, label='$I(x)=80x-0.5x^2$')
ax2_tw = ax.twinx()
ax2_tw.plot(x_v, IM_v, color=C_ROJO, lw=2, ls='--', label="$IM=80-x$")
ax2_tw.axhline(0, color='gray', ls=':', lw=1)
ax.axvline(80, color='gray', ls=':', lw=1)
ax.set_xlabel('x'); ax.set_ylabel('Ingreso', color=C_VERDE)
ax2_tw.set_ylabel('Ingreso marginal', color=C_ROJO)
ax.set_title('Ingreso Total e\nIngreso Marginal')

ax2=axes[1]
ax2.plot(L_v, Q_v, color=C_AZUL, lw=2.5, label='$Q(L)=50\\sqrt{L}$')
ax2_t = ax2.twinx()
ax2_t.plot(L_v, PMg_v, color=C_ROJO, lw=2, ls='--', label="$PMgL=25/\\sqrt{L}$")
ax2.set_xlabel('Trabajadores L'); ax2.set_ylabel('Producción Q', color=C_AZUL)
ax2_t.set_ylabel('Producto marginal', color=C_ROJO)
ax2.set_title('Función de Producción\ncon RMD')

ax3=axes[2]
ax3.plot(t_v, V_v, color=C_MORADO, lw=2.5, label='$V(t)=45000e^{-0.15t}$')
dV_v = -6750*np.exp(-0.15*t_v)
ax3_t = ax3.twinx()
ax3_t.plot(t_v, dV_v, color=C_ROJO, lw=2, ls='--', label="$V'(t)=-6750e^{-0.15t}$")
ax3.set_xlabel('Años t'); ax3.set_ylabel('Valor V(t)', color=C_MORADO)
ax3_t.set_ylabel('Tasa depreciación', color=C_ROJO)
ax3.set_title('Depreciación Exponencial')

plt.suptitle('Reglas Básicas de Derivación — Aplicaciones Económicas', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.savefig('fig_c03_reglas_basicas.png', bbox_inches='tight')
plt.show()
```

## Ejemplo 3.3 — Regla del Producto y del Cociente
### Ingreso marginal con demanda decreciente y costo medio

```python
# EJEMPLO 3.3 — Regla del producto y cociente
x = sp.Symbol('x', positive=True)

# Regla del producto: I(x) = p(x)·x = (100-2x)·x
p_fun = 100 - 2*x
I_fun = p_fun * x
IM_prod = sp.diff(I_fun, x)
x_opt = sp.solve(IM_prod, x)[0]
print('REGLA DEL PRODUCTO: I(x) = (100-2x)·x')
print(f'  IM(x) = {sp.expand(IM_prod)}')
print(f'  IM=0 → x*={x_opt}, P*={float(p_fun.subs(x,x_opt))}')

# Curva de aprendizaje: c(t) = (2t+5)·e^(-0.1t)
t = sp.Symbol('t', positive=True)
c_t = (2*t + 5)*sp.exp(-0.1*t)
dc_t = sp.diff(c_t, t)
t_crit = sp.solve(dc_t, t)[0]
print(f'\nCURVA DE APRENDIZAJE: c(t) = (2t+5)e^(-0.1t)')
print(f'  c\'(t) = {sp.simplify(dc_t)}')
print(f'  Punto crítico: t={float(t_crit):.1f}')

# Regla del cociente: costo medio
C_fun = x**3 - 6*x**2 + 15*x + 100
C_med = C_fun / x
dCmed = sp.diff(C_med, x)
x_min_cm = sp.solve(dCmed, x)
print(f'\nCOSTO MEDIO: C̄(x) = C(x)/x')
print(f'  C̄\'(x) = {sp.simplify(dCmed)}')
print(f'  Mínimo del costo medio en x≈{[float(r) for r in x_min_cm if r.is_real and float(r)>0]}')

x_v = np.linspace(0.1, 50, 400)
I_v  = (100-2*x_v)*x_v
IM_v = 100-4*x_v
t_v  = np.linspace(0, 20, 400)
c_v  = (2*t_v+5)*np.exp(-0.1*t_v)
Cm_v = (x_v**3-6*x_v**2+15*x_v+100)/x_v
dCm_v= 2*x_v - 6 - 100/x_v**2

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
ax=axes[0]
ax.plot(x_v, I_v, color=C_VERDE, lw=2.5, label='$I(x)=(100-2x)x$')
ax2_tw=ax.twinx()
ax2_tw.plot(x_v, IM_v, color=C_ROJO, lw=2, ls='--', label='$IM=100-4x$')
ax2_tw.axhline(0, color='gray', ls=':', lw=1)
ax.axvline(25, color='gray', ls=':', lw=1)
ax.set_xlabel('x'); ax.set_ylabel('Ingreso', color=C_VERDE)
ax2_tw.set_ylabel('IM', color=C_ROJO)
ax.set_title('Ingreso Total y Marginal\n(Regla del Producto)')

ax2=axes[1]
ax2.plot(t_v, c_v, color=C_AZUL, lw=2.5, label='$c(t)=(2t+5)e^{-0.1t}$')
t_c = float(t_crit)
ax2.scatter([t_c],[float((2*t_c+5)*np.exp(-0.1*t_c))],color=C_ROJO,s=100,zorder=6,
    label=f't crítico = {t_c:.1f}')
ax2.set_xlabel('Meses t'); ax2.set_ylabel('Costo unitario')
ax2.set_title('Curva de Aprendizaje\n(Producto: polinomio × exp)'); ax2.legend()

ax3=axes[2]
ax3.plot(x_v, Cm_v, color=C_MORADO, lw=2.5, label='$\\bar{C}(x)$')
ax3_t=ax3.twinx()
ax3_t.plot(x_v, dCm_v, color=C_ROJO, lw=2, ls='--', label="$\\bar{C}'(x)$")
ax3_t.axhline(0, color='gray', ls=':', lw=1)
ax3.set_xlabel('x'); ax3.set_ylabel('Costo medio', color=C_MORADO)
ax3_t.set_ylabel("Derivada costo medio", color=C_ROJO)
ax3.set_title('Costo Medio y su Derivada\n(Regla del Cociente)')

plt.suptitle('Regla del Producto y del Cociente', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.savefig('fig_c03_producto_cociente.png', bbox_inches='tight')
plt.show()
```

## Ejemplo 3.4 — Regla de la Cadena
### Composición de funciones en modelos económicos

```python
# EJEMPLO 3.4 — Regla de la cadena
x = sp.Symbol('x')
t = sp.Symbol('t', positive=True)

# (a) h(x) = (3x²+2x+1)^5
ha = (3*x**2 + 2*x + 1)**5
dha = sp.diff(ha, x)
print(f'h(x)=(3x²+2x+1)^5  →  h\'(x) = {sp.factor(dha)}')

# (b) C(x) = e^(0.02x²)
x_pos = sp.Symbol('x', positive=True)
Cb = sp.exp(0.02*x_pos**2)
dCb = sp.diff(Cb, x_pos)
print(f'C(x)=e^(0.02x²)  →  C\'(x) = {dCb}')

# (c) D(t) = ln(2t²+5)
Dc = sp.ln(2*t**2 + 5)
dDc = sp.diff(Dc, t)
print(f'D(t)=ln(2t²+5)  →  D\'(t) = {dDc}')

# (d) P(t) = 1000 e^(-0.05t²)
Pd = 1000*sp.exp(-0.05*t**2)
dPd = sp.diff(Pd, t)
print(f'P(t)=1000e^(-0.05t²)  →  P\'(t) = {sp.simplify(dPd)}')

# Gráficas
x_v = np.linspace(-1, 1, 400)
ha_v = (3*x_v**2+2*x_v+1)**5
dha_v = np.array([float(dha.subs(x, xv)) for xv in x_v])

t_v = np.linspace(0.1, 5, 400)
Dc_v = np.log(2*t_v**2+5)
dDc_v = 4*t_v/(2*t_v**2+5)

t_v2 = np.linspace(0, 8, 400)
Pd_v = 1000*np.exp(-0.05*t_v2**2)
dPd_v = -100*t_v2*np.exp(-0.05*t_v2**2)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
ax=axes[0]
ax.plot(x_v, ha_v, color=C_AZUL, lw=2.5, label='$h(x)=(3x^2+2x+1)^5$')
ax2_t=ax.twinx()
ax2_t.plot(x_v, dha_v, color=C_ROJO, lw=2, ls='--', label="$h'(x)$")
ax.set_xlabel('x'); ax.set_ylabel('h(x)', color=C_AZUL)
ax2_t.set_ylabel("h'(x)", color=C_ROJO)
ax.set_title('$(3x^2+2x+1)^5$\nRegla de la cadena')

ax2=axes[1]
ax2.plot(t_v, Dc_v, color=C_VERDE, lw=2.5, label='$D(t)=\\ln(2t^2+5)$')
ax2_t2=ax2.twinx()
ax2_t2.plot(t_v, dDc_v, color=C_ROJO, lw=2, ls='--', label="$D'(t)$")
ax2.set_xlabel('t'); ax2.set_ylabel('D(t)', color=C_VERDE)
ax2_t2.set_ylabel("D'(t)", color=C_ROJO)
ax2.set_title('$\\ln(2t^2+5)$\nDifusión tecnológica')

ax3=axes[2]
ax3.plot(t_v2, Pd_v, color=C_NARANJA, lw=2.5, label='$P(t)=1000e^{-0.05t^2}$')
ax3_t=ax3.twinx()
ax3_t.plot(t_v2, dPd_v, color=C_ROJO, lw=2, ls='--', label="$P'(t)$")
ax3.set_xlabel('t'); ax3.set_ylabel('P(t)', color=C_NARANJA)
ax3_t.set_ylabel("P'(t)", color=C_ROJO)
ax3.set_title('$1000e^{-0.05t^2}$\nCurva de Gauss')

plt.suptitle('Regla de la Cadena — Aplicaciones', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.savefig('fig_c03_cadena.png', bbox_inches='tight')
plt.show()
```

## Ejemplo 3.5 — Derivación Implícita
### Isocuanta, curva de indiferencia y equilibrio de mercado

```python
# EJEMPLO 3.5 — Derivación implícita: isocuantas y curvas de indiferencia
K, L_s = sp.symbols('K L', positive=True)
x_s, y_s = sp.symbols('x y', positive=True)

# Isocuanta: K^0.4 * L^0.6 = Q0 → TMST = dK/dL = -(0.6/0.4)*(K/L)
# Por derivación implícita:
alpha, beta = 0.4, 0.6
TMST = -(beta/alpha)*(K/L_s)
print('ISOCUANTA: K^0.4 · L^0.6 = Q₀')
print(f'  TMST = dK/dL = -(β/α)·(K/L) = {TMST}')
print(f'  En K=16, L=81: TMST = {float(TMST.subs([(K,16),(L_s,81)])):.4f}')

# Curva de indiferencia: U(x,y) = x^0.3 * y^0.7 = U0 → TMS = -(0.3/0.7)*(y/x)
TMS = -(0.3/0.7)*(y_s/x_s)
print(f'\nCURVA DE INDIFERENCIA: x^0.3 · y^0.7 = U₀')
print(f'  TMS = dy/dx = -(0.3/0.7)·(y/x)')
print(f'  En (9,4): TMS = {float(TMS.subs([(x_s,9),(y_s,4)])):.4f}')

# Círculo: x² + y² = 25
# dy/dx = -x/y
print(f'\nCIRCUNFERENCIA: x²+y²=25')
print(f'  dy/dx = -x/y')
print(f'  En (3,4): dy/dx = {-3/4}')

# Isocuantas
Q0_vals = [100, 150, 200]
L_rng = np.linspace(1, 100, 400)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
ax=axes[0]
for Q0, col in zip(Q0_vals, [C_AZUL,C_VERDE,C_ROJO]):
    K_iso = (Q0 / L_rng**0.6)**(1/0.4)
    ax.plot(L_rng, K_iso, color=col, lw=2, label=f'Q₀={Q0}')
ax.set_xlim(0,100); ax.set_ylim(0,200)
ax.set_xlabel('Trabajo L'); ax.set_ylabel('Capital K')
ax.set_title('Isocuantas Cobb-Douglas\n$K^{0.4}L^{0.6}=Q_0$'); ax.legend()

# Curvas de indiferencia
ax2=axes[1]
U0_vals = [6, 10, 15]
x_rng = np.linspace(0.5, 50, 400)
for U0, col in zip(U0_vals, [C_AZUL,C_VERDE,C_ROJO]):
    y_ci = (U0/x_rng**0.3)**(1/0.7)
    ax2.plot(x_rng, y_ci, color=col, lw=2, label=f'U₀={U0}')
ax2.set_xlim(0,50); ax2.set_ylim(0,80)
ax2.set_xlabel('Bien x'); ax2.set_ylabel('Bien y')
ax2.set_title('Curvas de Indiferencia\n$x^{0.3}y^{0.7}=U_0$'); ax2.legend()

# Círculo con tangente
ax3=axes[2]
theta = np.linspace(0, 2*np.pi, 300)
ax3.plot(5*np.cos(theta), 5*np.sin(theta), color=C_AZUL, lw=2.5, label='$x^2+y^2=25$')
# Tangente en (3,4): y-4 = (-3/4)(x-3)
x_tan = np.linspace(-1, 7, 100)
y_tan = 4 + (-3/4)*(x_tan-3)
ax3.plot(x_tan, y_tan, color=C_ROJO, lw=2, ls='--', label='Tangente en (3,4)')
ax3.scatter([3],[4],color='black',s=100,zorder=6)
ax3.annotate('(3,4)\nm=-3/4',xy=(3,4),xytext=(1,2),fontsize=10,
    arrowprops=dict(arrowstyle='->',color='gray'))
ax3.set_aspect('equal'); ax3.set_xlabel('x'); ax3.set_ylabel('y')
ax3.set_title('Derivación Implícita\n$x^2+y^2=25$'); ax3.legend(fontsize=9)

plt.suptitle('Derivación Implícita', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.savefig('fig_c03_implicita.png', bbox_inches='tight')
plt.show()
```

## Ejemplo 3.6 — Derivadas de Orden Superior y Análisis Marginal
### Costo cúbico: economías y deseconomías de escala

```python
# EJEMPLO 3.6 — Derivadas de orden superior y análisis marginal
x = sp.Symbol('x', positive=True)

# Costo cúbico del libro
C = 0.002*x**3 - 0.6*x**2 + 80*x + 1500
C1 = sp.diff(C, x)   # Costo marginal
C2 = sp.diff(C1, x)  # Segunda derivada
C3 = sp.diff(C2, x)  # Tercera derivada

x_inflexion = sp.solve(C2, x)[0]
CM_min = float(C1.subs(x, x_inflexion))

print('FUNCIÓN DE COSTO: C(x) = 0.002x³ - 0.6x² + 80x + 1500')
print(f"  C'(x)  = {C1}  (Costo marginal)")
print(f"  C''(x) = {C2}  (Tasa de cambio del CM)")
print(f"  C'''(x)= {C3}")
print(f"  Punto de inflexión del CM: x = {x_inflexion}")
print(f"  Costo marginal mínimo: C'({x_inflexion}) = {CM_min}")
print(f"  Para x < {x_inflexion}: C''<0 → CM decreciente (economías de escala)")
print(f"  Para x > {x_inflexion}: C''>0 → CM creciente (deseconomías de escala)")

# Maximización de beneficio: π = I - C con I=120x
I_f = 120*x
pi = I_f - C
pi1 = sp.diff(pi, x)
pi2 = sp.diff(pi1, x)
x_opt = sp.solve(pi1, x)
x_opt_pos = [float(r) for r in x_opt if r.is_real and float(r) > 0]
print(f"\nBeneficio óptimo:")
for x_o in x_opt_pos:
    pi_val = float(pi.subs(x, x_o))
    pi2_val= float(pi2.subs(x, x_o))
    tipo = 'MÁXIMO' if pi2_val < 0 else 'MÍNIMO'
    print(f"  x={x_o:.2f}: π={pi_val:.2f}, π''={pi2_val:.2f} → {tipo}")

x_v = np.linspace(0, 200, 500)
C_v  = 0.002*x_v**3 - 0.6*x_v**2 + 80*x_v + 1500
CM_v = 0.006*x_v**2 - 1.2*x_v + 80
C2_v = 0.012*x_v - 1.2

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
ax=axes[0]
ax.plot(x_v, C_v, color=C_AZUL, lw=2.5, label="$C(x)$")
ax.axvline(100, color='gray', ls=':', lw=1)
ax.set_xlabel('x'); ax.set_ylabel('C(x)'); ax.set_title('Costo Total Cúbico'); ax.legend()

ax2=axes[1]
ax2.plot(x_v, CM_v, color=C_VERDE, lw=2.5, label="$C'(x)$ Costo marginal")
ax2.axvline(100, color='gray', ls=':', lw=1)
ax2.scatter([100],[CM_min],color=C_ROJO,s=100,zorder=6,label=f"Mínimo CM={CM_min:.0f}")
ax2.fill_between(x_v, CM_v, CM_min, where=(x_v<100), alpha=0.2, color=C_VERDE, label='Economías')
ax2.fill_between(x_v, CM_v, CM_min, where=(x_v>100), alpha=0.2, color=C_ROJO, label='Deseconomías')
ax2.set_xlabel('x'); ax2.set_ylabel("C'(x)"); ax2.set_title('Costo Marginal'); ax2.legend(fontsize=9)

ax3=axes[2]
ax3.plot(x_v, C2_v, color=C_NARANJA, lw=2.5, label="$C''(x)$")
ax3.axhline(0, color='black', lw=0.8)
ax3.axvline(100, color='gray', ls=':', lw=1)
ax3.fill_between(x_v, C2_v, 0, where=(x_v<100), alpha=0.2, color=C_VERDE, label="C''<0 (economías)")
ax3.fill_between(x_v, C2_v, 0, where=(x_v>100), alpha=0.2, color=C_ROJO, label="C''>0 (deseconomías)")
ax3.set_xlabel('x'); ax3.set_ylabel("C''(x)")
ax3.set_title('Segunda Derivada del Costo'); ax3.legend(fontsize=9)

plt.suptitle('Derivadas de Orden Superior — Economías y Deseconomías de Escala', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.savefig('fig_c03_orden_superior.png', bbox_inches='tight')
plt.show()
```
